```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df = pd.read_csv('chipotle_usa.csv')
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Total Lemak Kalori berdasarkan Category

```python
category = df.groupby('Menu Category').count()[['TotalFatCalories']].sort_values(by='TotalFatCalories',ascending=True).reset_index()
category = category.rename(columns={'TotalFatCalories':'CategoryOfFat'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=category['Menu Category'], y=category['CategoryOfFat'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi Kolestrol

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Cholesterol'])
```

# distribusi Jumlah Kalori

```python
plt.figure(figsize=(15,5))
sns.distplot(df['TotalCalories'])
```

# seleksi fitur

```python
features = ['TotalFats','Cholesterol','Sodium','Carbohydrates','DietaryFiber','Sugar','Protein']
x = df[features]
y = df['TotalCalories']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score ,'%')
```

# membuat inputan model regresi linier

```python
input_data = np.array([[9,0,600,0,3,0,8]])

prediction = lr.predict(input_data) 
print('Estimasi jumlah kalori (kCal) :', prediction)
```

# save model

```python
import pickle

filename = 'estimasi_calChipotle.sav'
pickle.dump(lr,open(filename,'wb'))
```
